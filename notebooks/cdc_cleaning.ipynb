```python
cdc['region'] = ""
for ind,row in cdc.iterrows():
    if row._STATE == 'Connecticut' or row._STATE == 'Puerto Rico' or row._STATE == 'Maine' or row._STATE == 'Massachusetts' or row._STATE == 'New Hampshire' or row._STATE == 'Rhode Island' or row._STATE == 'Vermont' or row._STATE == 'New Jersey' or row._STATE == 'New York' or row._STATE == 'Pennsylvania':
        cdc.at[ind, 'region'] = 'Northeast'
    elif row._STATE == 'Illinois' or row._STATE == 'Indiana' or row._STATE == 'Michigan' or row._STATE == 'Ohio' or row._STATE == 'Wisconsin' or row._STATE == 'Iowa' or row._STATE == 'Kansas' or row._STATE == 'Minnesota' or row._STATE == 'Missouri' or row._STATE == 'Nebraska' or row._STATE == 'North Dakota' or row._STATE == 'South Dakota':
        cdc.at[ind, 'region'] = 'Midwest'
    elif row._STATE == 'Arizona' or row._STATE == 'Guam' or row._STATE == 'Colorado' or row._STATE == 'Idaho' or row._STATE == 'Montana' or row._STATE == 'Nevada' or row._STATE == 'New Mexico' or row._STATE == 'Utah' or row._STATE == 'Wyoming' or row._STATE == 'Alaska' or row._STATE == 'California' or row._STATE == 'Hawaii' or row._STATE == 'Oregon' or row._STATE == 'Washington':
        cdc.at[ind, 'region'] = 'West'
    elif row._STATE == np.nan:
        cdc.at[ind, 'region'] = np.nan
    else:
        cdc.at[ind,'region'] = 'South'
```

```python
cdc['trans'] = ""
for ind,row in cdc.iterrows():
    if row.TRNSGNDR == 'Yes, male to female' or row.TRNSGNDR == 'Yes, female to male' or row.TRNSGNDR == 'Yes, gender nonconforming':
        cdc.at[ind, 'trans'] = 'Transgender'
    elif row.TRNSGNDR == 'No' or row.TRNSGNDR =="Don't Know/Not Sure" or row.TRNSGNDR == 'Refused':
        cdc.at[ind,'trans'] = 'Cisgender'
    else:
        cdc.at[ind,'trans'] = 'Unknown'
```

```python
cdc['gender'] = ""
for ind,row in cdc.iterrows():
    if row.TRNSGNDR == 'Yes, female to male':
        cdc.at[ind,'gender'] = 'Male'
    elif row.TRNSGNDR == 'Yes, male to female':
        cdc.at[ind,'gender'] = 'Female'
    elif row.TRNSGNDR == 'Yes, gender nonconforming':
        cdc.at[ind,'gender'] = 'Gender nonconforming'
    else:
        cdc.at[ind,'gender'] = row._SEX
```

```python
cdc['sexual orientation'] = ""
for ind,row in cdc.iterrows():
    if row.SOMALE == 'Gay' or row.SOFEMALE == 'Lesbian or Gay':
        cdc.at[ind,'sexual orientation'] = 'Gay'
    elif row.SOMALE == 'Straight' or row.SOFEMALE == 'Straight':
        cdc.at[ind,'sexual orientation'] = 'Straight'
    elif row.SOMALE == 'Bisexual' or row.SOFEMALE == 'Bisexual':
        cdc.at[ind,'sexual orientation'] = 'Bisexual'
    elif row.SOMALE == 'Something else' or row.SOFEMALE == 'Something else':
        cdc.at[ind,'sexual orientation'] = 'Something else'
```

```python
cdc['education'] = ""
for ind,row in cdc.iterrows():
    if row._EDUCAG == 'Did not graduate High School' or row._EDUCAG == 'Graduated High School':
        cdc.at[ind,'education'] = 'No post-secondary education'
    elif row._EDUCAG == 'Attended College or Technical School' or row._EDUCAG == 'Graduated from College or Technical School':
        cdc.at[ind,'education'] = 'Post-secondary education'
```

```python
cdc['employment'] = ""
for ind,row in cdc.iterrows():
    if row.EMPLOY1 == 'Employed for Wages' or row.EMPLOY1 == 'Self-employed':
        cdc.at[ind,'employment'] = 'Employed'
    elif row.EMPLOY1 == 'Out of work for 1 year or more' or row.EMPLOY1 == 'Out of work for 1 year or more':
        cdc.at[ind,'employment'] = 'Unemployed'
```

```python
cdc['income']= ""
for ind,row in cdc.iterrows():
    if row._INCOMG == "Less than 15,000" or row._INCOMG == "15,000 to less than 25,000" or row._INCOMG == '25,000 to less than 35,000':
        cdc.at[ind,'income'] = 'Less than $35,000'
    elif row._INCOMG == '35,000 to less than 50,000' or row._INCOMG == '50,000 or more':
        cdc.at[ind,'income'] = '$35,000 or more'
```
